```python
import requests
import gradio as gr

# Function to get weather information
def get_weather(city_name):
    api_key = "a2addd9e5bf95c3dda8fd35fa6b43da6"
    base_url = "http://api.openweathermap.org/data/2.5/weather?"
    complete_url = f"{base_url}q={city_name}&appid={api_key}"
    response = requests.get(complete_url)

    if response.status_code == 200:
        data = response.json()
        weather_description = data["weather"][0]["description"]
        temperature_kelvin = data["main"]["temp"]
        temperature_celsius = temperature_kelvin - 273.15
        humidity = data["main"]["humidity"]
        wind_speed = data["wind"]["speed"]

        result = f"""
        <div style="background-color: #f2f2f2; padding: 20px; border-radius: 10px;">
            <h2 style="color: #333;">Weather in {city_name}</h2>
            <p><strong>Description:</strong> {weather_description}</p>
            <p><strong>Temperature:</strong> {temperature_celsius:.2f}°C</p>
            <p><strong>Humidity:</strong> {humidity}%</p>
            <p><strong>Wind Speed:</strong> {wind_speed} m/s</p>
        </div>
        """
        return result
    else:
        return "City not found or API request failed."

# Create a Gradio interface with custom HTML and CSS
iface = gr.Interface(
    fn=get_weather,
    inputs="text",
    outputs=gr.outputs.HTML(),
    live=True,
    capture_session=True,
    theme="huggingface",
    css="""
        body {
            background-color: #f0f0f0;
            font-family: Arial, sans-serif;
            text-align: center;
        }
        .output { margin-top: 20px; }
    """
)

iface.launch()
```

```python
import torch
from transformers import AutoModelForMaskedLM, AutoTokenizer

# Load the pre-trained BERT model and tokenizer
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForMaskedLM.from_pretrained(model_name)

# Function to perform text correction using BERT
def text_correction(input_text):
    # Tokenize the input text
    tokens = tokenizer.tokenize(input_text)

    # Mask one token in the input text
    masked_index = tokens.index("[MASK]")
    tokens[masked_index] = "[MASK]"

    # Convert tokens back to text
    text = " ".join(tokens)

    # Encode the text and predict the masked word
    input_ids = tokenizer.encode(text, return_tensors="pt")
    with torch.no_grad():
        predictions = model(input_ids)

    # Get the predicted word
    predicted_index = torch.argmax(predictions.logits[0, masked_index]).item()
    predicted_word = tokenizer.decode([predicted_index])

    # Replace the masked token with the predicted word
    tokens[masked_index] = predicted_word

    # Convert tokens back to corrected text
    corrected_text = " ".join(tokens)

    return corrected_text

# Example usage
input_text = "I hav a good boook at home."
corrected_text = text_correction(input_text)
print("Original Text:", input_text)
print("Corrected Text:", corrected_text)
```
